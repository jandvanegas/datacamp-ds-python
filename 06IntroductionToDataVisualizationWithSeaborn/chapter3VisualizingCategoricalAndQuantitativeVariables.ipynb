```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
survey_data = pd.read_csv('../data/06IntroductionToDataVisualizationWithSeaborn/young-people-survey-responses.csv')
student_data = pd.read_csv('../data/06IntroductionToDataVisualizationWithSeaborn/student-alcohol-consumption.csv')
```

```python
# Create column subplots based on age category
sns.catplot(y="Internet usage", data=survey_data,
            kind="count", col='Age')

# Show plot
plt.show()
```

```python
# Create a bar plot of interest in math, separated by gender

sns.catplot(x='Gender', y='Mathematics', data=survey_data, kind='bar')


# Show plot
plt.show()
```

```python
# Turn off the confidence intervals
sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar",
            order=["<2 hours",
                   "2 to 5 hours",
                   "5 to 10 hours",
                   ">10 hours"],
                   ci=None)

# Show plot
plt.show()
```

```python
# Specify the category ordering
study_time_order = ["<2 hours", "2 to 5 hours",
                    "5 to 10 hours", ">10 hours"]

# Create a box plot and set the order of the categories
sns.catplot(x='study_time', y='G3', data=student_data,
order=study_time_order, kind='box')




# Show plot
plt.show()
```

```python
# Create a box plot with subgroups and omit the outliers

sns.catplot(x='internet', y='G3', data=student_data,
            kind='box', sym='', hue='location')




# Show plot
plt.show()
```

```python
# Extend the whiskers to the 5th and 95th percentile
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=0.5)

# Show plot
plt.show()
```

```python
# Set the whiskers at the min and max values
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[5, 95])

# Show plot
plt.show()
```

```python
# Set the whiskers at the min and max values
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[0, 100])

# Show plot
plt.show()
```

```python
sns.catplot(x="famrel", y="absences",
			data=student_data,
            kind="point",
            capsize=0.2)

# Show plot
plt.show()
```

```python
# Remove the lines joining the points
sns.catplot(x="famrel", y="absences",
			data=student_data,
            kind="point",
            capsize=0.2, join=False)

# Show plot
plt.show()
```

```python
# Import median function from numpy
from numpy import median

# Plot the median number of absences instead of the mean
sns.catplot(x="romantic", y="absences",
			data=student_data,
            kind="point",
            hue="school",
            ci=None,
            estimator=median)

# Show plot
plt.show()
```
